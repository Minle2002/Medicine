```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Medicine_Details.csv')
df.head()
```

```python
df = df.drop(df.columns[[1, 4, 5, 6, 7, 8]], axis = 1)
```

```python
df['Uses'] = df['Uses'].str.split('Treatment of')
df = df.explode('Uses')
```

```python
df['Uses'] = df['Uses'].str.split('Prevention of')
df = df.explode('Uses')
```

```python
df = df.replace('', np.nan).dropna(axis=1, how='all')
df = df.dropna()
```

```python
def add_commas(text):
    result = ''
    for char in text.strip():
        if char.isupper() and char != ' ':
            result += ', ' + char
        else:
            result += char
    return result
```

```python
df['Uses'] = df['Uses'].apply(add_commas)
df['Side_effects'] = df['Side_effects'].apply(add_commas)
```

```python
df['Uses'] = df['Uses'].str.lstrip(', ')
df = df.explode('Uses')
df['Side_effects'] = df['Side_effects'].str.lstrip(', ')
df = df.explode('Side_effects')
```

```python
df['Uses'] = df['Uses'].str.replace(' ,', ',')
df['Side_effects'] = df['Side_effects'].str.replace(' ,', ',')
```

```python
df['Uses'] = df['Uses'].str.replace(r'\(.*?\)', '', regex=True)
```
